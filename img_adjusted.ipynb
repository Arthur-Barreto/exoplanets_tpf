# Flatten the image

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
# carregar os dados de cada categoria, e plotar um histograma do tamanho do flatten da imagem

current_path = os.getcwd()
print(f"path autal={current_path}")
```

```python
candidate_path = os.path.join(current_path, "CANDIDATE")
confirmed_path = os.path.join(current_path, "CONFIRMED")
false_positive_path = os.path.join(current_path, "FALSE POSITIVE")
```

## Análise para `CANDIDATE`

```python
candidate_flatten = []

for file in os.listdir(candidate_path):
    file = os.path.join(candidate_path, file)
    if os.path.isfile(file):
        img = np.load(file)
        loaded_img = img["matrix_img"]
        flatten = loaded_img.flatten()
        # save the flatten size
        candidate_flatten.append(len(flatten))
    
```

```python
print(min(candidate_flatten), max(candidate_flatten))
```

```python
import scipy.stats as stats

plt.hist(
    candidate_flatten,
    bins=int(np.sqrt(len(candidate_flatten))),
    density=True,
    alpha=0.6,
    color="g",
)
kde = stats.gaussian_kde(candidate_flatten)
plt.plot(np.linspace(0, 2300, 100), kde(np.linspace(0, 2300, 100)), color="g")
plt.title("Candidate")
# plot a red line for the mean
plt.axvline(np.mean(candidate_flatten), color="r", linestyle="--")
#  write the mean value
plt.text(
    np.mean(candidate_flatten) + 20,
    0.0005,
    f"Mean: {np.mean(candidate_flatten):.2f}",
    color="r",
)
plt.show()
```

```python
# print the value <= 100 and the porcentange of the total
print(
    f"<= 100: {len([x for x in candidate_flatten if x <= 100])} - {len([x for x in candidate_flatten if x <= 100])/len(candidate_flatten)*100:.2f}%"
)
```
